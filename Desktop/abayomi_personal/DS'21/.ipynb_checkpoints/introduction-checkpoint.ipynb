#### Basic Operators

```python
# Arithmetic Operators (% Modulus, addition, subtraction)
tunde = 4
femi = tunde +5
print(femi)
# Comparison (Relational) Operators
tunde == femi
tunde > femi
femi != tunde
# Assignment Operators
femi = 34
femi += 3
# Logical Operators
femi == 37 and tunde == 34
# Membership Operators
fav_foods=[2,3,4,5]
4 in fav_foods
```

#### Decision Making


```python
# if conditional statement
a = 43
b = 23
if a >= b:
    print("a is greater than b")
    
    
# when we have more than a condition
if a >= b:
    print("a is greater than b")
elif a<=b:
    print("a is less than b")
    
# We can use if to close up, to handle any case whose consition is not stated
if a >= b:
    print("a is greater than b")
elif a<=b:
    print("a is less than b")
else:
    print("we are not sure of your condition")
```
