### Feature Scaling

```python
# age is a continous value and ML algorithm are very sensitive to that.
# Before we scale down, let's look out for outliers using box plot

sns.boxplot(df['age'])
```

```python
# Our data seems within a reasonable range
# lET'S SCALE DOWN OUR AGE
# There are diffent scaling methods
from sklearn.preprocessing import MinMaxScaler
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
```

```python
from sklearn.preprocessing import StandardScaler
max_encoder = MinMaxScaler()
```

```python
df_scaled = pd.DataFrame(max_encoder.fit_transform(df),columns = df.columns)
```

```python
df_scaled
```

```python
# Since we are predicting, we will definitely need a y, for every x(s) there is a y or y(s)
X = df_scaled.drop('class', axis=1)
y = df_scaled['class']
```

### Data Splitting - There are different statistical tools for splitting
##### We are doing random sampling

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state=42)
```

```python
X_train
```

#### We are ready to train our model and make predictions

```python
# Linear Regression
from sklearn.linear_model import LinearRegression
lgr = LinearRegression()
lgr.fit(X_train,y_train)
```

```python
sub_test_x = X_test.iloc[:5]
sub_test_y = y_test.iloc[:5]
```

```python
sub_test_x
```

```python
sub_test_y
```

#### Let's Predict and Compare

```python
lgr_prediction = lgr.predict(sub_test_x)
```

```python
lgr_prediction
```

#### Checking accuracy of model

```python
model_prediction = lgr.predict(X_test)
from sklearn.metrics import mean_absolute_error
lgr_mse = mean_absolute_error(model_prediction, y_test)
lgr_mse

# The absolute error is the absolute value of the difference between the forecasted value and the actual value
```

### Let's use decision Tree

```python
from sklearn.tree import DecisionTreeRegressor
tree_reg = DecisionTreeRegressor()
tree_reg.fit(X_train,y_train)
```

```python
import tensorflow as tf
import tensorflow.keras as keras
from keras.layers import Dense
```

```python
kernel_initializer='glorot_uniform'
ann_model1 = keras.models.Sequential([
        keras.layers.Dense(20, activation='relu',kernel_initializer='uniform', input_dim = X_train.shape[1]),
        keras.layers.Dense(20, activation='relu', kernel_initializer=kernel_initializer),
        keras.layers.Dense(1),
])
print(ann_model1.summary())

# loss and optimizer

loss = keras.losses.CategoricalCrossentropy(from_logits=True)
optim = keras.optimizers.Adam(lr=0.001)

metrics = ['accuracy']

ann_model1.compile(loss = 'categorical_crossentropy', optimizer=optim, metrics=metrics)



batch_size = 3
epochs = 40
model = ann_model1.fit(X_train, y_train,validation_data=(X_test, y_test), batch_size = batch_size, epochs=epochs,verbose=2)
scores = ann_model1.evaluate(X_test,y_test)
print("\n%s: %.2f%%" % (ann_model1.metrics_names[1], scores[1]*100))

ann_model1.save("model.h5")
```
